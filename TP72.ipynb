```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x=pd.DataFrame(iris.data)
x.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']
```

```python
wcss_list= [] #Initializing the list for the values of WCSS
```

```python
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1, 10), wcss_list)
plt.title('Le graphe de la la méthode du coude (Elbow)')
plt.xlabel('Nombre de clusters(k)')
plt.ylabel('Liste des wcss')
plt.show()
```

```python
model=KMeans(n_clusters=3)
```

```python
model.fit(x)
```

```python
colormap=np.array(['Red','green','blue','yellow','cyan','black'])
plt.scatter(x.Petal_Length, x.Petal_width,c=colormap[model.labels_],s=20)
```
