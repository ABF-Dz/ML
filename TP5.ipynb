# Partie 1

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
```

```python
X=2*np.random.rand(100,1)
```

```python
y = np.sin(X[:,0]) + 0.0 * np.random.randn(100)
```

```python
plt.scatter(X, y, s=30)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
y_predicted = reg.predict(X)
```

```python
rmse = mean_squared_error(y, y_predicted)
```

```python
print("Pente : " ,reg.coef_)
print("Ordonnée à l'origine : ", reg.intercept_)
print("Racine carrée de l'erreur quadratique moyenne : ", rmse)
```

```python
plt.scatter(X,y, s=50)
plt.xlabel("$X$")
plt.ylabel("$y$")
x_min, x_max = plt.xlim()
nx = 100
xx = np.linspace(x_min, x_max, nx).reshape(-1,1)
plt.plot(xx,reg.predict(xx), color='red', label="Régression linéaire")
plt.title("Régression linéaire unidimensionnelle")
plt.legend()
plt.show()
```

# Partie 2

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=8)
x_poly=poly_reg.fit_transform(X)
x_poly
lin_reg_model2 = LinearRegression()
```

```python
lin_reg_model2.fit(x_poly, y)
y_pred2 = lin_reg_model2.predict(x_poly)
y_pred2
```

```python
x_grid = np.arange(min(X), max(X), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(X, y, color='green')
plt.plot(x_grid, lin_reg_model2.predict(poly_reg.fit_transform(x_grid)), color="blue")
plt.title("Polynomial Regression")
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
rmse = mean_squared_error(y, y_pred2)
print("Racine carrée de l'erreur quadratique moyenne : ", rmse)
```
