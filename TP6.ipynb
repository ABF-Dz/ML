# Partie 1

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import sklearn
```

```python
data = pd.read_csv('customer_purchases.csv')
```

```python
print(data)
print("Keys of data: \n{}".format(data.keys()))
```

```python
plt.scatter(data['Age'],data['Salary'], alpha=1, s=100, c=data['Purchased'])
plt.xlabel('Age'), plt.ylabel('Salary'), plt.title('Scatter plot on our Data')
```

```python
X = data.iloc[:,0:1]
y = data.iloc[:, 2]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=109)
```

```python
from sklearn import svm
cls1 = svm.SVC(kernel="linear")
```

```python
cls1.fit(X_train,y_train)
```

```python
pred1 = cls1.predict(X_test)
```

```python
from sklearn import metrics
print("Precision_1:", metrics.precision_score(y_test,y_pred=pred1))
```
