```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x=np.array([6,8,10,14,18]).reshape((-1,1))
y=np.array([7,9,13,17,18])
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=8)
x_poly=poly_reg.fit_transform(x)
x_poly
lin_reg_model2=LinearRegression()
lin_reg_model2.fit(x_poly,y)
y_pred2=lin_reg_model2.predict(x_poly)
y_pred2
```

```python
plt.scatter(x,y,color="green",label='')
plt.plot(x,y_pred2,color="blue",label='')
plt.title("Poly Reg")
plt.xlabel("§x§")
plt.ylabel("§y§")
plt.show()
```

```python
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color="green")
plt.plot(x_grid,lin_reg_model2.predict(poly_reg.fit_transform(x_grid)),color='blue')
plt.title("Poly Reg")
plt.xlabel("§x§")
plt.ylabel("§y§")
plt.show()
```

```python
rmse=mean_squared_error(y,y_pred2)
print("rmse=",rmse)
```
